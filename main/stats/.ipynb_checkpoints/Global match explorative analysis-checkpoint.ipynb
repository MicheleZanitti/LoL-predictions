### Load statistics from 100 random matches

```python
# extract 100 random gameids
gameids = db.games.find()
storedids = []
maxids = 100
i = 0
for g in gameids:
    if(i<maxids and bool(random.getrandbits(1))):
        storedids.append(g['gameId'])
        i += 1
    elif(i>=maxids):
        break
```

```python
# save global statistics about all 100 games, we want to find correlation between the win/lose and the other events.
globaldfstats = []

for gameid in storedids:
    game = db.games.find({'gameId': gameid})[0]
    statsparticipant = []
    for p in game['participants']:

        statsparticipant.append(p['stats'])


    dfstats = pd.DataFrame(statsparticipant)
    globaldfstats.append(dfstats)
```

```python
globaldf = pd.concat(globaldfstats).reset_index(drop=True)
```

```python
# interesting variables to look at
globalvars = ['assists',
             'champLevel',
             'damageDealtToObjectives',
             'damageDealtToTurrets',
             'damageSelfMitigated',
             'deaths',
             'doubleKills',
             'firstBloodKill',
             'firstInhibitorAssist',
             'firstInhibitorKill',
             'firstTowerAssist',
             'firstTowerKill',
             'goldEarned',
             'goldSpent',
             'inhibitorKills',
             'timeCCingOthers',
             'totalDamageDealt',
             'totalDamageDealtToChampions',
             'totalDamageTaken',
             'totalHeal',
             'totalMinionsKilled',
             'totalTimeCrowdControlDealt',
             'totalUnitsHealed',
             'tripleKills',
             'trueDamageDealt',
             'trueDamageDealtToChampions',
             'trueDamageTaken',
             'turretKills',
             'visionScore',
             'visionWardsBoughtInGame',
             'wardsKilled',
             'wardsPlaced',
             'win',
             'physicalDamageDealt',
             'physicalDamageDealtToChampions',
             'physicalDamageTaken',
             'killingSprees',
             'kills',
             'largestCriticalStrike',
             'largestKillingSpree',
             'largestMultiKill',
             'longestTimeSpentLiving',
             'magicDamageDealt',
             'magicDamageDealtToChampions',
             'magicalDamageTaken',
             'neutralMinionsKilled',
             'neutralMinionsKilledEnemyJungle',
             'neutralMinionsKilledTeamJungle',
             'participantId',
             'pentaKills']


# more interesting variables to look at
filteredvars = ['assists',
                'champLevel',
                'damageDealtToObjectives',
                'damageDealtToTurrets',
                'deaths',
                'doubleKills',
                'firstTowerAssist',
                'firstTowerKill',
                'goldEarned',
                'goldSpent',
                'inhibitorKills',
                'killingSprees',
                'kills',
                'largestKillingSpree',
                'win']
```

### Matches mean statistics

```python
# grouping stats by match result and compute their mean
groupedglobaldf = globaldf[filteredvars].groupby(['win']).mean().T
```

```python
# change boolean columns type to integer
filtered_data = globaldf[filteredvars].copy()
boolcolumns = list(filtered_data.select_dtypes(include=['bool']).columns)
for col in boolcolumns:
    filtered_data[col] = filtered_data[col].astype(int)
```

```python
# scale to frequency for each variable
groupedglobaldf = groupedglobaldf.T
filtereddf = groupedglobaldf.copy()
for col in groupedglobaldf.columns:
    filtereddf[col] = filtereddf[col]/filtereddf[col].sum()
# show data
filtereddf.T
```

```python
# plot the variables frequencies
filtereddf.T.plot.bar(figsize=(18,8), title='Mean global statistics per result')
stats.saveImage('varStatisticsResult')
```

### Variable correlations with game result

```python
# plot the correlation matrix
plt.figure(figsize=(18,18))
plt.title('Variable correlations with match result')
sns.heatmap(globaldf[filteredvars].corr(),cbar=False,cmap='coolwarm',annot=True)
# stats.saveImage('heatmapWinCorrelation') # uncomment to save the figure
```
